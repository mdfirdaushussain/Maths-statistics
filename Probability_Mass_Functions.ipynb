```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# One Dice Rolling

```python
arr=np.array(np.random.randint(1,7,size=1000))
print(arr)
```

```python
df=(pd.Series(arr).value_counts()/pd.Series(arr).value_counts().sum()).sort_index()
```

```python
x=df.index
y=df.values
```

```python
plt.bar(x,y)
```

### Cumulative distribution function

```python
df.cumsum().plot(kind='bar')
```

# Two Dice Rolling

```python
a=np.array(np.random.randint(1,7,size=1000))
b=np.array(np.random.randint(1,7,size=1000))
result=a+b
print(result)
print(len(result))
```

```python
df=(pd.Series(result).value_counts()/pd.Series(result).value_counts().sum()).sort_index()
print(pd.Series(result).value_counts(),'\n', pd.Series(result).value_counts().sum())
```

```python
plt.bar(df.index,df.values)
```

### Cumulative distribution function

```python
df.cumsum().plot(kind='bar')
```
